## Text Preprocessing Stemming Uing NLTK

Stemming: reduzir palavras às suas formas mais básicas

```python
## Classification problem
## Comments about a product are positive or negative
## Reviews ---> eat, eating, eaten
## Stemming ---> eat, eat, eat

words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

### Porter Stemmer

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

for wowrd in words:
    print(wowrd,"--->",ps.stem(wowrd))
```

```python
ps.stem("congatulation") ## misspelled word, major issue for stemming
```

### RegexStemmer class

We can easily implement Regular Expression Steamming algorithms. Remove all preffix and suffix that matches with the expression, only returning the regular expression

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingeating')
```

### Snowball stemmer
Return a better form of words than Poter Stemmer

```python
from nltk.stem import SnowballStemmer
```

```python
snowballsstemmer=SnowballStemmer('english')
```

```python
for word in words:
    print(word+"---->"+snowballsstemmer.stem(word))
```

```python
ps.stem("fairly"),ps.stem("sportingly")
```

```python
snowballsstemmer.stem("fairly"),snowballsstemmer.stem("sportingly")
```

```python
snowballsstemmer.stem('goes')
```
