## Encapsulation
Restringir acessos dentro de um objeto, prevenindo erros

Getters and Setters methods
Public, protected and private variables

Public -> Acesso direto ao atributo
Private -> Acesso apenas dentro da própria classe ou com setters e getters
Protected -> Não pode ser acessado de fora da classe, mas pode ser acessado por classes derivadas

```python
class Person:
    def __init__(self, name, age, gender):
        self.name = name ## public attribute
        self.__age = age  ## private attribute
        self._gender = gender ## protected attribute

    def get_age(self):
        return self.__age
    
    def set_age(self, age):
        if age >= 0:
            self.__age = age
        else:
            print("Invalid age")

person = Person("Lucas", 25, 'Male')

## Atributos públicos
print(person.name)       # Acessado direto
person.name = 'Maria'    # Tentativa de modificação direta
print(person.name)

## Atributos privados + getters and setters
## print(person.__age)  # Tentativa de acesso direto (gera erro)
print(person.get_age())  # Acessado via método público
person.set_age(30)      # Modificado via método público
print(person.get_age())  # Verificação da modificação

## Protected attribute
class Employee(Person):
    def get_gender(self):
        return self._gender

employee = Employee("Ana", 30, 'Female')
print(employee._gender)  # Tentativa de acesso direto (possível mas não recomendado)
print(employee.get_gender())  # Acessado via método da classe derivada
```
