## Abstraction
-> Esconder a complexidade de algo e apenas as features necess√°rias

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

    @abstractmethod
    def stop_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("Car engine started.")

    def stop_engine(self):
        print("Car engine stopped.")

Car().start_engine() # Dont need to create an object of Vehicle class as it is abstract
Car().stop_engine()
```
