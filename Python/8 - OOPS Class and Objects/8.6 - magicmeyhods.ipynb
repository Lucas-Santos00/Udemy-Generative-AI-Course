## Magic methods
São ganchos internos que o Python chama automaticamente para permitir que seus objetos se comportem como tipos nativos.
Eles habilitam operadores, funções internas, protocolos e integração com o ecossistema da linguagem.

__init__: Inicializa uma nova instancia da classe
__str__: Retorna uma representação do objeto como uma string
__repr__: Retorna uma string oficial que representa o objeto
__len__: Retorna o length do objeto
__getitem__: Recupera um item do container
__setitem__: Setar um item no container

```python
class Person():
    pass

person = Person()

print(person) ## Aqui está a saída do print(person) padrão
```

```python
## Basic methods

class Person():
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def __str__(self):
        return f"Person named {self.name} aged {self.age} years." ## Aqui estamos sobrescrevendo o método __str__, dando um novo comportamento a ele.
    
    def __repr__(self):
        return f"Person('{self.name}', {self.age})" ## Aqui estamos sobrescrevendo o método __repr__, dando um novo comportamento a ele.
    
person = Person("Lucas", 23)

print(person)
print(repr(person))
```
