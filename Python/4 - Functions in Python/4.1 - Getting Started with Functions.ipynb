Getting start with functions

```python
## Funções são blocos de código que realizam uma ação específica
## Syntax
def function_name(parameters='Default Value'): ## Default value é opcional
    """" Docstring: Descrição da função """
    expression = "Hello, World!"  # Ação a ser realizada
    print(expression)  # Exibe a mensagem na tela
    return True


function_name()  # Chamada da função
```

```python
## Positional and Keyword Arguments
# Parâmetros posicionais são passados na ordem em que são definidos na função.
# Parâmetros nomeados são passados usando o nome do parâmetro, permitindo que sejam especificados em qualquer ordem.
# Exemplo de função com parâmetros posicionais e nomeados variáveis:
def function_name(*args, **kwargs): ## ** retrepresenta parâmetros nomeados variáveis - exemplo, kwargs = {'key1': 'value1', 'key2': 'value2'}
    """
    Docstring: Descrição da função
    args: Parâmetros posicionais variáveis
    kwargs: Parâmetros nomeados variáveis
    """
    print("Positional arguments:", args)
    print("Keyword arguments:", kwargs)
    return True

function_name(1, 2, 3, key1="value1", key2="value2")  # Chamada da função com parâmetros variáveis
```

```python
## Return statement
# A instrução `return` é usada para retornar um valor de uma função.
# Se não for especificado, a função retorna `None` por padrão.
def add_numbers(a, b):
    """
    Docstring: Adiciona dois números e retorna o resultado
    a: Primeiro número
    b: Segundo número
    """
    return a + b, 'teste'  # Retorna a soma dos números  - Aceitando multiplos valores de retorno

rst = add_numbers(5, 3)  # Chamada da função e atribuição do resultado a variáveis
print("Mensagem:", rst)  # Exibe a mensagem retornada pela função
```
