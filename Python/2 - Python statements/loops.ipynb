# Loops In Python

```python
## For loop
# for é um dos loops que utiliza uma sequência de valores para iterar, passando com cada valor da sequência.

for i in range(1, 6, 2): # (número inicial(Opcional), número final, passo(Opcional))
    print(i)

print("---")

for i in range(10, 1, -1): # (número inicial(Opcional), número final, passo(Opcional))
    print(i)
```

```python
# While loop
# while é um loop que executa enquanto uma condição for verdadeira.

count = 0
while count < 5:
    print(count)
    count += 1
```

```python
# Loop control statements
# break: interrompe o loop.
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
# continue: pula para a próxima iteração do loop.
for i in range(10):
    if i == 5:
        continue
    print(i)
```

```python
# pass
# pass: não faz nada, mas é necessário para manter a sintaxe correta.
for i in range(10):
    if i == 5:
        pass
    print(i)
```

```python
# nested loops
# loops aninhados são loops dentro de outros loops.
for i in range(3):
    for j in range(2):
        print(f"i: {i}, j: {j}")
```
