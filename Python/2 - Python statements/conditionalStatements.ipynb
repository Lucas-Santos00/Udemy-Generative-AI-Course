```python
## If statements
# If é um bloco de código que é executado se uma condição for verdadeira.

age = 18

if age >= 18:
    print("Você é maior de idade.")
elif age == 17:
    print('Este ano você não pode votar, mas no próximo ano poderá.')
else: ## Sempre execitado se a condição for falsa
    print("Você é menor de idade.")
```

```python
## Exercício ano bissexto
year = 2028

if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0):
    print(f"{year} é um ano bissexto.")
else:
    print(f"{year} não é um ano bissexto.")
```
