Tuples

```python
## Tuplas são uma coleção ordenada de itens de itens imutáveis, são iguais a listas - mas imutáveis

## Criando uma tupla
empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst=list() ## Criando uma lista vazia
tpl=tuple() ## Criando uma tupla vazia
```

```python
numbers=tuple([1,2,3,4,5,6]) ## lista convertida em tupla
numbers2=list((1,2,3,4,5,6)) ## tupla convertida em lista

print(type(numbers))
print(type(numbers2))
```

```python
## Tuplas podem receber qualquer tipo de dado, inclusive listas e tuplas
mixed_tuple=(1,2,3,4,5,6,"Python",True,[1,2,3],(1,2,3))
print(mixed_tuple) 
```

```python
## Acessando os valores de uma tupla
print(mixed_tuple[0]) ## Acessando o primeiro valor
print(mixed_tuple[-1]) ## Acessando o ultimo valor
```

```python
## Operações com tuplas
print(mixed_tuple[0:3]) ## Acessando os 3 primeiros valores
print(mixed_tuple[3:]) ## Acessando os valores a partir do quarto valor
print(mixed_tuple[:3]) ## Acessando os 3 primeiros valores
print(mixed_tuple[::2]) ## Acessando os valores pares
print(mixed_tuple[::-1]) ## Acessando os valores em ordem inversa
```

```python
## Concatenação de tuplas
tuple1=(1,2,3)
tuple2=(4,5,6)
tuple3=tuple1+tuple2
print(tuple3) ## (1, 2, 3, 4, 5, 6)
```

```python
## Natureza imutável das tuplas
# mixed_tuple[0]=10 ## Isso irá gerar um erro, pois tuplas são imutáveis
# TypeError: 'tuple' object does not support item assignment
```

```python
## Métodos de tuplas
print(mixed_tuple.count(1)) ## Contando quantas vezes o valor 1 aparece
print(mixed_tuple.index(1)) ## Retorna o índice do primeiro valor 1 encontrado
print(mixed_tuple.index("Python")) ## Retorna o índice do primeiro valor "Python"   
```

```python
## Packing and Unpacking de Tuplas
# Packing
packed_tuple = (1, 2, 3, "Python", True)
print(packed_tuple)
# Unpacking
a, b, c, d, e = packed_tuple
print(a, b, c, d, e)
```

```python
## unpacking de tuplas com * (asterisco)
a, *b, c = packed_tuple
print(a)  # 1
print(b)  # [2, 3, 'Python']  
print(c)  # True
```

```python
## Tuplas aninhadas
nested_tuple = (1, 2, (3, 4), (5, 6))
print(nested_tuple)  # (1, 2, (3, 4), (5, 6))
print(nested_tuple[2])  # (3, 4)
print(nested_tuple[2][0])  # 3  
```
