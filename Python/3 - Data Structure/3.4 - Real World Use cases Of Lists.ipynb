Real world use case

```python
to_do_list=['Buy Groceries', 'Clean the house', 'Finish homework', 'Pay bills']

## Adding to tasks
to_do_list.append('Schedule meeting')
to_do_list.append('Go For Run')

## Removing from tasks
to_do_list.remove('Clean the house')

## Checking if a task is in the list
if 'Pay bills' in to_do_list:
    print("You have to pay bills today.")

## Remeaning tasks
for task in to_do_list:
    print(f'-{task}')
```
