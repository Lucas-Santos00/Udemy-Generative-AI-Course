```python
## Basic Syntax Rules In pYthon
## Case Sensitivity

name='Lucas'
Name='Lucas2'

print(name)
print(Name)
# The above code shows that Python is case-sensitive.
```

```python
## Identation

age = 32

if age > 18:
    print("You are an adult")
else:   
    print("You are a minor")
# The above code shows that Python uses indentation to define blocks of code.
```

```python
## Line continuation

total = 1 + 2 + 3 + \
        4 + 5 + 6 + \
        7 + 8 + 9
print(total)
# The above code shows that Python allows line continuation using the backslash (\).
```

```python
## Multiple statements on a single line
x = 1; y = 2; z = 3
print(x, y, z)
```

```python
## Understand Semantics In Python
# Variable assignment

x = 10 ## Type is inferred as int automatically
y = 3.14 ## Type is inferred as float automatically
z = "Hello, World!" ## Type is inferred as str automatically

type_x = type(x)
type_y = type(y)
type_z = type(z)
print(f"x is of type: {type_x}")
print(f"y is of type: {type_y}")
print(f"z is of type: {type_z}")
# The above code shows that Python automatically infers the type of a variable based on the value assigned to it.
```

```python
## Type inference
# Python automatically infers the type of a variable based on the value assigned to it.

variable=10
print(f"The type of variable is: {type(variable)}")
variable='Lucas'
print(f"The type of variable is: {type(variable)}")
```
